```python
import pandas as pd
import matplotlib.pyplot as plt

# Membuat file CSV langsung di Colab
data = """Hari,Pengunjung
Senin,120
Selasa,135
Rabu,150
Kamis,160
Jumat,140
Sabtu,200
Minggu,180
"""
with open("perpustakaan.csv", "w") as f:
    f.write(data)

# Baca data
df = pd.read_csv("perpustakaan.csv")

# Plot jumlah pengunjung
plt.figure(figsize=(10,6))
plt.plot(df["Hari"], df["Pengunjung"], marker="o", linestyle="-", color="green")

plt.title("Jumlah Pengunjung Perpustakaan Selama Seminggu", fontsize=16)
plt.xlabel("Hari", fontsize=12)
plt.ylabel("Jumlah Pengunjung", fontsize=12)
plt.grid(True)

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Data perpustakaan (pengunjung total per hari)
data = {
    "Hari": ["Senin","Selasa","Rabu","Kamis","Jumat","Sabtu","Minggu"],
    "Pengunjung": [120,135,150,160,140,200,180]
}
df = pd.DataFrame(data)

# Ubah jadi bentuk matriks (1 minggu x 1 variabel)
df_matrix = df.pivot_table(values="Pengunjung", index="Hari", aggfunc="sum")

# Plot heatmap
plt.figure(figsize=(6,5))
sns.heatmap(df_matrix, annot=True, cmap="YlOrRd", fmt="d")

plt.title("Heatmap Jumlah Pengunjung Perpustakaan", fontsize=16)
plt.ylabel("Hari", fontsize=12)
plt.xlabel("Jumlah Pengunjung", fontsize=12)

plt.show()
```

```python
import plotly.express as px
import pandas as pd

# Data fiktif perpustakaan
data = {
    'Hari': ['Senin', 'Selasa', 'Rabu', 'Kamis', 'Jumat', 'Sabtu', 'Minggu'],
    'Pengunjung': [120, 135, 150, 160, 140, 200, 180],
    'Durasi': [1.5, 1.7, 2.0, 2.2, 2.1, 2.5, 2.3]  # durasi rata-rata (jam)
}
df = pd.DataFrame(data)

# Scatter plot interaktif
fig = px.scatter(
    df,
    x='Hari',
    y='Pengunjung',
    size='Durasi',       # ukuran titik = durasi kunjungan
    color='Pengunjung',  # warna = jumlah pengunjung
    hover_data=['Durasi'],
    title='Jumlah Pengunjung dan Durasi Kunjungan Perpustakaan (Interaktif)'
)

fig.show()
```

```python
import matplotlib.pyplot as plt

# Data
hari = ["Senin", "Selasa", "Rabu", "Kamis", "Jumat"]
anak = [30, 35, 40, 38, 32]
remaja = [40, 45, 50, 48, 42]
dewasa = [50, 55, 60, 58, 52]

# Membuat figure lebih besar
plt.figure(figsize=(8, 5))

# Beberapa variasi garis (kategori pengunjung)
plt.plot(hari, anak, marker='o', color='blue', linestyle='-', label='Anak')
plt.plot(hari, remaja, marker='s', color='red', linestyle='--', label='Remaja')
plt.plot(hari, dewasa, marker='^', color='green', linestyle='-.', label='Dewasa')

# Menambahkan judul dan label
plt.title("Jumlah Pengunjung Perpustakaan Berdasarkan Kategori", fontsize=14)
plt.xlabel("Hari")
plt.ylabel("Jumlah Pengunjung")

# Anotasi pada titik puncak
plt.annotate("Puncak Remaja", xy=("Rabu", 50), xytext=("Kamis", 55),
             arrowprops=dict(facecolor='black', shrink=0.05))

# Tambahan grid dan legend
plt.grid(True, linestyle=':', color='gray')
plt.legend()
plt.show()
```

```python
import plotly.express as px
import pandas as pd

# Data perpustakaan
df = pd.DataFrame({
    "Hari": ["Senin", "Selasa", "Rabu", "Kamis", "Jumat"],
    "Anak": [30, 35, 40, 38, 32],
    "Dewasa": [50, 55, 60, 58, 52],
    "Total": [80, 90, 100, 96, 84]
})

# Line Chart interaktif (Anak vs Dewasa)
fig1 = px.line(df, x="Hari", y=["Anak", "Dewasa"], markers=True,
               title="Jumlah Pengunjung Anak & Dewasa (Line Chart)")
fig1.show()

# Scatter interaktif (Total pengunjung per hari)
fig2 = px.scatter(df, x="Hari", y="Total", size="Total", color="Total",
                  title="Total Pengunjung Perpustakaan (Scatter)")
fig2.show()

# Bar Chart interaktif (Pengunjung Anak)
fig3 = px.bar(df, x="Hari", y="Anak", color="Anak",
              title="Jumlah Pengunjung Anak (Bar Chart)")
fig3.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Data perpustakaan fiktif
data = {
    "hari": ["Senin", "Selasa", "Rabu", "Kamis", "Jumat", "Sabtu", "Minggu"],
    "pengunjung": [120, 135, 150, 160, 140, 200, 180],
    "durasi": [1.5, 1.7, 2.0, 2.2, 2.1, 2.5, 2.3],
    "kategori": ["weekday", "weekday", "weekday", "weekday", "weekday", "weekend", "weekend"],
    "gender_dom": ["Male", "Female", "Female", "Male", "Male", "Female", "Male"]
}
df = pd.DataFrame(data)

plt.figure(figsize=(12, 6))

# Histogram distribusi jumlah pengunjung
plt.subplot(1, 2, 1)
sns.histplot(df["pengunjung"], bins=10, kde=True, color="green")
plt.title("Distribusi Jumlah Pengunjung")

# Scatterplot hubungan pengunjung vs durasi
plt.subplot(1, 2, 2)
sns.scatterplot(x="pengunjung", y="durasi", hue="kategori", style="gender_dom", data=df, s=100)
plt.title("Pengunjung vs Durasi Rata-rata")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data perpustakaan
hari = ["Senin", "Selasa", "Rabu", "Kamis", "Jumat"]
pengunjung = [120, 135, 150, 160, 140]  # jumlah pengunjung
durasi = [1.5, 1.7, 2.0, 2.2, 2.1]     # rata-rata durasi kunjungan (jam)

plt.figure(figsize=(12, 5))

# Line Chart (jumlah pengunjung per hari)
plt.subplot(1, 3, 1)
plt.plot(hari, pengunjung, marker='o', color='blue', label='Pengunjung')
plt.title("Line Chart: Pengunjung")
plt.legend()
plt.grid(True)

# Bar Chart (jumlah pengunjung)
plt.subplot(1, 3, 2)
plt.bar(hari, pengunjung, color='orange')
plt.title("Bar Chart: Pengunjung")

# Scatter Plot (pengunjung vs durasi)
plt.subplot(1, 3, 3)
plt.scatter(pengunjung, durasi, color='green', s=100)
plt.title("Scatter: Pengunjung vs Durasi")
plt.xlabel("Jumlah Pengunjung")
plt.ylabel("Durasi (jam)")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data perpustakaan
hari = ["Senin", "Selasa", "Rabu", "Kamis", "Jumat"]
pengunjung = [120, 135, 150, 160, 140]      # jumlah pengunjung
dipinjam = [80, 90, 100, 110, 95]          # buku dipinjam
dikembalikan = [70, 85, 95, 105, 100]      # buku dikembalikan

plt.figure(figsize=(8,5))

plt.plot(hari, pengunjung, marker='o', color='hotpink', linestyle='-', linewidth=2, label="Pengunjung")
plt.plot(hari, dipinjam, marker='s', color='purple', linestyle='--', linewidth=2, label="Buku Dipinjam")
plt.plot(hari, dikembalikan, marker='^', color='deeppink', linestyle='-.', linewidth=2, label="Buku Dikembalikan")

plt.title("Aktivitas Perpustakaan (SeninJumat)")
plt.xlabel("Hari")
plt.ylabel("Jumlah")
plt.grid(True, linestyle=":")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

# Data hari
hari = np.linspace(0, 5, 100)  # SeninJumat (5 hari)
cmap = plt.get_cmap('jet', 7)  # 7 warna untuk 7 minggu

fig, ax = plt.subplots(figsize=(8, 6))

# Simulasi data: jumlah pengunjung tiap minggu
for i, minggu in enumerate(np.linspace(0, 2, 7)):
    ax.plot(hari, 100 + minggu*20 + np.sin(hari*2)*10, color=cmap(i), label=f"Minggu {i+1}")

# Tambahkan colorbar
sm = mpl.cm.ScalarMappable(cmap=cmap, norm=mpl.colors.Normalize(0, 2))
plt.colorbar(sm, ax=ax, ticks=np.linspace(0, 2, 7), label="Minggu")

plt.xlabel('Hari (SeninJumat)')
plt.ylabel('Jumlah Pengunjung')
plt.title("Perbandingan Jumlah Pengunjung Perpustakaan Selama 7 Minggu")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data fiktif: jumlah buku dipinjam (10 kategori  10 hari)
np.random.seed(42)
data = np.random.randint(20, 100, size=(10, 10))

# Label kategori buku dan hari
kategori = [f"Kategori {i}" for i in range(1, 11)]
hari = [f"Hari {i}" for i in range(1, 11)]

fig, ax = plt.subplots(figsize=(8, 6))
cax = ax.imshow(data, cmap='coolwarm')

# Menambahkan colorbar horizontal
plt.colorbar(cax, orientation='horizontal', label='Jumlah Buku Dipinjam')

# Menambahkan label
ax.set_xticks(np.arange(len(hari)))
ax.set_yticks(np.arange(len(kategori)))
ax.set_xticklabels(hari, rotation=45, ha="right")
ax.set_yticklabels(kategori)

plt.title("Heatmap Jumlah Buku Dipinjam di Perpustakaan (10 Hari)")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10, 8))  # Buat figure baru
fig.suptitle('Visualisasi Data Perpustakaan')  # Judul utama

# Subplot 1: Pertumbuhan pengunjung (logaritmik)
ax1 = fig.add_subplot(221)
ax1.plot([np.log(n+1)*50 for n in range(1, 10)], marker='o')
ax1.set_title("Pertumbuhan Pengunjung (Log)")

# Subplot 2: Pola kunjungan (sinusoidal)
ax2 = fig.add_subplot(222, facecolor='lightyellow')
hari = np.linspace(0, 10, 100)
ax2.plot(np.sin(hari) * 50 + 100, color='blue')
ax2.set_title("Pola Kunjungan Harian")

# Subplot 3: Jumlah buku dipinjam (acak)
ax3 = fig.add_subplot(223)
ax3.plot(np.random.randint(50, 150, 10), marker='o', linestyle='--', color='r')
ax3.set_title("Jumlah Buku Dipinjam (10 Hari)")

# Subplot 4: Buku paling populer
ax4 = fig.add_subplot(224)
kategori = ['Fiksi', 'Sains', 'Sejarah', 'Teknologi']
jumlah = [120, 90, 75, 60]
ax4.bar(kategori, jumlah, color=['blue', 'orange', 'green', 'purple'])
ax4.set_title("Buku Paling Populer")

plt.tight_layout(rect=[0, 0, 1, 0.95])  # Atur layout
plt.show()
```
