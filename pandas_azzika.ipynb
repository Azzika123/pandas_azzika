```python
#metode dataframe.assign()
import pandas as pd

# Define a dictionary containing Students data
data = {'Nama': ['gyeonu', 'changmin', 'joohyuk', 'wooseok'],
        'Tinggi': [179, 176, 180, 185],
        'Kualifikasi': ['A', 'B', 'C', 'D']}
# Convert the dictionary into DataFrame
df = pd.DataFrame(data)
# using assign() and passing address list as parameter
df = df.assign(address = ['seoul', 'tokyo', 'busan', 'hanyang'])

print(df)
```

```python
#dataframe.loc()
import pandas as pd

# dictionary containing Students data
data = {'Name': ['Pandas', 'Geeks', 'for', 'Geeks'],
        'Height': [1, 2, 3, 4],
        'Qualification': ['A', 'B', 'C', 'D']}
# Convert the dictionary into DataFrame
df = pd.DataFrame(data)
# Using .loc[] to add a "Category" column
# based on height condition
df.loc[df['Height'] >= 3, 'Category'] = 'Tall'
df.loc[df['Height'] < 3, 'Category'] = 'Short'

# Observe the result
print(df)
```

```python
#GROUPING DATA
#1. Kelompokkan data berdasarkan satu kunci:
import pandas as pd
data1 = {'Name':['louis', 'zayn', 'harry', 'nial',
                 'liam', 'eunwoo', 'rafael', 'roger'],
        'Age':[30, 31, 32, 31,
               30, 25, 21, 22],
        'Address':['amerika', 'new york', 'los angeles', 'london',
                   'paris', 'korea', 'jakarta', 'tokyo'],
        'Qualification':['Msc', 'MA', 'MCA', 'Phd',
                         'B.Tech', 'B.com', 'Msc', 'MA']}
df = pd.DataFrame(data1)
print(df)
df.groupby('Name')
print(df.groupby('Name').groups)
```

```python
#2. Transformasi
import pandas as pd

# bikin DataFrame dari dictionary
df2 = pd.DataFrame({
    'Name': ['baim', 'indra', 'aqsa', 'ikra',
             'sari', 'kila', 'zikzik', 'halim'],
    'Age': [19, 17, 18, 21, 19, 15, 20, 22],
    'Address': ['new york', 'los angeles', 'chicago', 'houston',
                'philadelphia', 'phoenix', 'san atonio', 'san diego'],
    'Qualification': ['Msc', 'MA', 'MCA', 'Phd',
                      'B.Tech', 'B.com', 'Msc', 'MA'],
    'Score': [24, 30, 27, 26, 31, 22, 20, 38]
})

# groupby berdasarkan Name
grp2 = df2.groupby('Name')

# fungsi standardisasi
sc = lambda x: (x - x.mean()) / x.std()

# transform kolom Age
df2['Age_scaled'] = grp2['Age'].transform(sc)

print(df2)
```

```python
#filtrasi
grp2 = df2.groupby('Name')
grp2.filter(lambda x: len(x) >= 2)
```
